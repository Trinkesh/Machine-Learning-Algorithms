# Logistic Regression on Social Network dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Avik-Jain/100-Days-Of-ML-Code/master/datasets/Social_Network_Ads.csv")
data.head()
```

### Data sepraton into dependent and independent

```python
X = data.iloc[:,2:4]
y = data.iloc[:,4]
```

```python
X
```

```python
y
```

## Splite The  Data Usning Train Test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
X_train
```

```python
y_train
```

Standardize the features age and estimated salary

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
```

```python
regression = regression.fit(X_train,y_train)
```

```python
regression
```

```python
X_train
```

```python
X_test
```

```python
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
```

```python
cf =classification_report(y_test,y_pred)
cf
```
