## Naive Baise algo with iris dataset

```python
#import essintial libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import datasets
```

```python
#load dataset from sklearn pre datasets 
iris = datasets.load_iris()
```

```python
#key elements of datasets
iris.keys()
```

### Split data into dependent and independent 

```python
X = iris.data
```

```python
y = iris.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state  = 0)
```

```python
X_train
```

```python
X_test
```

```python
#we are using naive bayes classification algo
from sklearn.naive_bayes import GaussianNB
naive = GaussianNB()
from sklearn import metrics
```

```python
#fit data 
nb = naive.fit(X_train,y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_pred,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_pred,y_test)
```

```python
metrics.accuracy_score(y_pred,y_test)
```
