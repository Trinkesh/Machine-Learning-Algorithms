```python
#import the essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#the training full data
#which contains dependent feature an output value Saleprice

X_full = pd.read_csv("https://raw.githubusercontent.com/demidafonichev/housing-prices-competition-for-kaggle-learn-users/master/data/train.csv")
X_full.head(3)
```

```python
#the testing data
#data which has no sale price or no dependent features
X_test_full  = pd.read_csv("https://raw.githubusercontent.com/demidafonichev/housing-prices-competition-for-kaggle-learn-users/master/data/test.csv")
X_test_full.head(3) 
```

```python
#lets obtain target and predected 
X_full.columns
```

```python
#the dependent feature
y = X_full["SalePrice"]
y
```

```python
#select the features we are working in 
features =['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
#make a copy 
X = X_full[features].copy()
#features selected from X_full training data
X_test = X_test_full[features].copy()
#features selected from X_test testing data
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train_test_split from sklearn and define testing and training ration by define the size of
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8,test_size = 0.2,random_state = 1)
```

```python
#to get overview off splited data
X_train.head()
```

```python
#define few model for  RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
model1 = RandomForestRegressor(n_estimators=50,random_state=0)
model2 = RandomForestRegressor(n_estimators=100,random_state = 0)
model3 = RandomForestRegressor(n_estimators=100,criterion='mse',random_state = 0)
model4 = RandomForestRegressor(n_estimators=200,min_samples_split=20,random_state = 0)
model5 = RandomForestRegressor(n_estimators=200,max_depth = 7,random_state = 0)
#put every model in list variable
models = [model1,model2,model3,model4,model5]
```

create a simple function to perform number.or say 5models of  RandomForestRegressor algoritham 
To select the best model out of the five, we define a function score_model() below.
This function returns the mean absolute error (MAE)
from the validation set. Recall that the best model will obtain the lowest MAE. 

```python
#for defining the mean absolute errors
from sklearn.metrics import mean_absolute_error
```

```python
def score_model(model, X_t=X_train, X_v=X_test, y_t=y_train, y_v=y_test):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_test,preds)
```

```python
for i in range(0,len(models)):
    mae = score_model(models[i])
    print("model %d mae: %d"%(i+1,mae))
```

```python
#model 3 have lees mae than every other model
```
