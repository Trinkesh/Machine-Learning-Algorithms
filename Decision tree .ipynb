```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data  = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X  = data.iloc[:,:-1]
```

```python
X
```

```python
y = data["salary_more_then_100k"]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_company = LabelEncoder()
l_job = LabelEncoder()
l_degree = LabelEncoder()
```

```python
X["encoded_company"] = l_company.fit_transform(X["company"])
X["encoded_job"] = l_job.fit_transform(X["job"])
X["encoded_degree"] = l_degree.fit_transform(X["degree"])
```

```python
X
```

```python
X = X.drop(["company","job","degree"],axis = 1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=0)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(X_train,y_train)
```

```python
model
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_pred,y_test)
cm
```
